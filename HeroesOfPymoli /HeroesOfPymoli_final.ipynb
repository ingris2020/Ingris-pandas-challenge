```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
#import collections 
```

```python
# Read the CSV into a Pandas DataFrame
purchase_data = "Resources/purchase_data.csv"

purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#total number of players 
total_players = len(purchase_data_df["SN"].value_counts())
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Basic Calculation 
unique_items = len((purchase_data_df["Item ID"]).unique())
average_price = (purchase_data_df["Price"]).mean()
number_of_purchases = (purchase_data_df["Purchase ID"]).count()
total_revenue = (purchase_data_df["Price"]).sum()

#Data Frame to hold results
summary_df = pd.DataFrame({"Unique Items":[unique_items],
                           "Average Price":[average_price], 
                           "Number of Purchases": [number_of_purchases], 
                           "Total Revenue": [total_revenue]})

# Update numbers 
pd.options.display.float_format="${:,.2f}".format

#Display DataFrame
summary_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender demographics  
#Count and total per gender
gender_count = purchase_data_df.groupby('Gender')
player_count= gender_count.nunique()["SN"]
#Percentage
gender_percent=player_count/total_players
gender_percent=gender_percent.map("{:,.2%}".format)
gender_percent
```

```python
#Gender Demographics
gender_demo = pd.merge(player_count,gender_percent, on="Gender")
gender_demographics =gender_demo.rename(columns={"SN_x":"Total_Players","SN_y":"Percentage" })
#Sortting Data
gender_demographics = gender_demographics.sort_values(by="Percentage", ascending=False)
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_revenue = purchase_data_df.groupby(['Gender'])["Price"].sum()
pd.options.display.float_format="${:,.2f}".format
gender_average_purchase = purchase_data_df.groupby(['Gender'])['Price'].mean()
gender_purchase_count=purchase_data_df.groupby(['Gender'])["Purchase ID"].count()
average_player_price = gender_revenue/player_count
average_player_price
```

```python
#Data Frame to hold purchasing Analysis (Gender)
gender_summary_df = pd.DataFrame({"Purchase Count":gender_purchase_count,
                           "Average Purchase Price":gender_average_purchase, 
                           "Total Purchase Value": gender_revenue,
                           "Avg Total Purchase per Player": average_player_price})

# Setting index to "gender"
gender_summary_df.index.name = "Gender"

#Display DataFrame
gender_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held  
bins = [0, 9, 14, 19, 24, 29, 34, 39, 99]

# Create the names for the four bins
age_group = ["<10", "10-14", "15-19", "20-24", "25-29,", "30-34", "35-39", "40+"]

#Add an extra column to view the bins created for the ages
purchase_data_df["Age_Group"]=pd.cut(purchase_data_df["Age"], bins, labels=age_group)
purchase_data_df
```

```python
#Age demographics  
age_purchase_data_df = purchase_data_df.loc[:, ["Gender", "SN", "Age", "Age_Group"]]
age_purchase_data_df = age_purchase_data_df.drop_duplicates()

items_age= age_purchase_data_df["Age_Group"].value_counts()
items_age

items_percent=items_age/total_players
items_percent=items_percent.map("{:,.2%}".format)
items_percent
```

```python
age_demographics_df = pd.DataFrame({"Total Count":items_age,
                                   "Percentage of Players":items_percent
                                   })
age_demographics_df = age_demographics_df.sort_index()
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Age Demographic Calculations
age_revenue = purchase_data_df.groupby(['Age_Group'])["Price"].sum()
age_average_purchase = purchase_data_df.groupby(['Age_Group'])['Price'].mean()
age_purchase_count=purchase_data_df.groupby(['Age_Group'])["Purchase ID"].count()
age_player_price = age_revenue/age_demographics_df["Total Count"]
age_player_price 
```

```python
 #Data Frame to hold purchasing Analysis (Age)
age_summary_df = pd.DataFrame({"Purchase Count":age_purchase_count,
                           "Average Purchase Price":age_average_purchase, 
                           "Total Purchase Value": age_revenue,
                           "Avg Total Purchase per Player": age_player_price})

# Setting index to "gender"
age_summary_df.index.name = "Age"

#Display DataFrame
age_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
average_purchase_price = purchase_data_df.groupby(['SN'])['Price'].mean()
revenue_by_player = purchase_data_df.groupby(['SN'])['Price'].sum()
count_by_player = purchase_data_df.groupby(['SN'])['Item ID'].count()

top_spender = pd.merge(average_purchase_price,revenue_by_player, on="SN")
top_spender = top_spender.rename(columns={"Price_x":"Average_Purchase","Price_y":"Total_Purchase" })

top_spender_merge =pd.merge(count_by_player, top_spender, on="SN")
top_spender_merge = top_spender_merge.sort_values(by="Total_Purchase", ascending=False)
top_spender_merge.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_list = purchase_data_df.groupby(["Item ID","Item Name"])
item_list2 = item_list[["SN"]].count().rename(columns={"SN":"Purchase Count"}) 

item_list2["Item Price"]= item_list["Price"].mean()
item_list2["Total Purchase Value"]= item_list["Price"].sum()
item_list2.sort_values(by="Purchase Count", ascending=False).head()
```
